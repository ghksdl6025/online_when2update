```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classifier = 'rf'
df = pd.read_excel('../result/Comparison/Overall result acc.xlsx',sheet_name=classifier)
```

```python
df = df.rename(columns = {'Unnamed: 0':'Classifier','Unnamed: 1': 'Log', 'Unnamed: 2':'Prefix'})
```

```python
df = df.drop(columns = ['classifier', 'Unnamed: 8']).drop(index = 69)
```

```python
df.columns.values
```

```python
dfs = df.loc[:, [ 'Train once Acc (Avg.)', 'Retrain variant Acc (Avg.)','Retrain label Acc (Avg.)', 'Retrain prodrift Acc (Avg.)', 'Retrain prefixtreeCDD Acc (Avg.)']]
```

```python
dfs = dfs.rename(columns = {'Retrain variant Acc (Avg.)': 'Variant',
                            'Retrain label Acc (Avg.)':'Label Dist.',
                            'Train once Acc (Avg.)':'Train once',
                            'Retrain prodrift Acc (Avg.)':'ProDrift', 
                            'Retrain prefixtreeCDD Acc (Avg.)':'PrefixtreeCDD'})
```

```python
columns_my_order = ['ProDrift','PrefixtreeCDD', 'Train once', 'Variant', 'Label Dist.']
fig, ax = plt.subplots()
fig.set_size_inches(10,4)
for position, column in enumerate(columns_my_order):
    ax.boxplot(dfs[column], positions=[position], vert=False, medianprops={'color':'red'}, widths=0.5)

# ax.set_xticks(range(position+1))
ax.set_yticklabels(columns_my_order, fontsize=12)
ax.set_xlim(-0.05,1.05)
ax.xaxis.grid()
# plt.title('Comparison of Accuracy Scores by Trigger Method Across Event Logs', fontsize = 15)
plt.xlabel('Accuracy', fontsize= 13)
# plt.show()
plt.savefig('../imgs/comparison_accuracy_trigger_boxplot2.pdf', bbox_inches='tight')


# fig.set_size_inches(10, 6)
# dfs.boxplot(vert=False)

# ylabel_list = []
# plt.subplots_adjust(left = 0.25)
# ylabel_list.append(i)
# plt.xlim(0,1.1)
    
# plt.set_yticklabels(ylabel_list)
# plt.show()
```
